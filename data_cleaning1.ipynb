### working on the salary column

```python
# By looking at the data I found that the salary column has some '-1' values, check how many

filt = (df['Salary Estimate'] == '-1')
df.loc[filt]
```

```python
# 214 rows has -1 value in the 'Salary Estimate' column, Drop/Remove these rows

df = df[df['Salary Estimate'] != '-1'] 
```

```python
df.shape
```

```python
df.head()
```

```python
# remove Glassdoor word from the Salary Estimate column. This code will remove the () and what inside
df['Salary Estimate'].str.replace('\(.*\)', '')

df['Salary Estimate'] = df['Salary Estimate'].str.replace('\(.*\)', '')
```

```python
# another way to do remove (Glassdoor est.) using lambda function, since the column has only 1 space 
# we will get 2 values in form of a list, the first element is the element we are interested in

# df['Salary Estimate'].apply(lambda x : x.split(' ')[0])
```

```python
# get ride of $ and K for the 'Salary Estimate' column
df['Salary Estimate'].str.replace('$','').str.replace('K','')

# Or

df['Salary Estimate'].apply(lambda x : x.replace('$','').replace('K',''))

df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x : x.replace('$','').replace('K',''))
```

```python
df.head()
```

```python
# make an extra column. the column will contain 2 values 1 for jobs with hourly rate and 0 for jobs that doesn't have
df['hourly'] = df['Salary Estimate'].apply(lambda x : 1 if 'per hour' in x.lower() else 0)
```

```python
# check if the code worked, by calling unique function on 'hourly' Series to see how many unique values we have 
df['hourly'].unique()
```

```python
# do the same as above for 
df['employer_provided'] = df['Salary Estimate'].apply(lambda x : 1 if 'employer provided salary' in x.lower() else 0)
```

```python
df['employer_provided'].unique()
```

```python
df['employer_provided'].value_counts()
```

```python
# remove 'per hour' and 'employer provided salary' strings from cells in column 'Salary Estimate'
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x : x.lower().replace('per hour','').replace('employer provided salary',''))
```

```python
# make columns for minimum salary and maximum salary

df['min_salary'] = df['Salary Estimate'].apply(lambda x : x.split('-')[0]).str.replace(':','')
df['max_salary'] = df['Salary Estimate'].apply(lambda x : x.split('-')[1])
```

```python
# making an average column
df['avg_salary'] = (df['max_salary'].astype(int) + df['min_salary'].astype(int))/2  
```

```python
df['avg_salary']
```

### working on the company name column

```python
# extract the company name text alone
df['company_txt'] = df.apply(lambda x : x['Company Name'] if x['Rating'] < 0 else x['Company Name'][:-4], axis = 1)
```

### working with the location column

```python
df['Location']
```

```python
# extract the state from the location column

#df['job_state'] = df['Location'].str.split(',', expand = True)[1]

# or

df['job_state'] = df['Location'].apply(lambda x : x.split(',')[1])
```

```python
df['same_state'] = df.apply(lambda x: 1 if x['Location'] == x['Headquarters'] else 0, axis = 1)
```

### working with Founded column

```python
# get the company age
df['age'] = df['Founded'].apply(lambda x : x if x < 1 else 2021 - x)
```
